### Data Exploration:

#### Checking the imbalance of target variable

```python
plt.figure(figsize=(12,6))
size = df['Class'].value_counts()
labels = 'Not Fraud', 'Fraud'
color = ['green', 'red']


plt.pie(size, colors = color, labels = labels,autopct = '.%2f%%')
plt.title('Fraud Class', fontsize = 20)
plt.axis('off')
plt.show()
```

#### The data of target variable is highly imbalanced

```python
Not_Fraud = df[df['Class']==0]
Fraud = df[df['Class']==1]
```

```python
Not_Fraud.Amount.describe()
```

```python
Fraud.Amount.describe()
```

```python
Fraud.Time.describe()
```

```python
Not_Fraud.Time.describe()
```

```python
plt.figure(figsize=(8,4))
group_labels = ['Not Fraud', 'Fraud']
sns.barplot(x='Time', data=df, hue='Class')
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
f.suptitle('Time of transaction vs Amount by class')
ax1.scatter(Fraud.Time, Fraud.Amount)
ax1.set_title('Fraud')
ax2.scatter(Not_Fraud.Time, Not_Fraud.Amount)
ax2.set_title('Not Fraud')
plt.xlabel('Time (in Seconds)')
plt.ylabel('Amount')
plt.show()
```

```python
plt.figure(figsize=(28,14))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

- There are some features which are negatively correlated to Target feature and some are nearly no correlation and some are positively correlated.
- No correlation among V1-V28 features.
- 'Amount' has positive correlation with fetaures V7, V20 and negative correlation with V2, V5
- 'Time' has negative correlation with fetaures V3, V11, V25
- Class has negative correlation with V17, V16, V14, V12, V10, V7, V3 and some positive correlation with V11, V4, V2
